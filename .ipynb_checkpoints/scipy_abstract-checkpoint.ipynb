```python
dt = 0.02
t = np.arange(0., 10., dt)
f = 5.0
y = np.sin(2*np.pi*f*t)+0.5*np.random.randn(t.size)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(t, y)
```

```python
Y = sp.fftpack.fft(y)
freqs = sp.fftpack.fftfreq(Y.size, d=dt)
idx = np.where(freqs>=0)
plt.plot(freqs[idx], np.abs(Y[idx]))
```

```python
pidx = (freqs < 4.5)|(freqs > 5.5)
nidx = (freqs < -5.5)|(freqs > -4.5)
YY = Y.copy()
YY[pidx & nidx] = 0.
yy = sp.fftpack.ifft(YY).real
```

```python
plt.plot(t, yy)
```
